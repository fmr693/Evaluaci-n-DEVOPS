### Ejercicio 20.
Dados los siguientes arrays:
```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
```
- Usa `np.where(cond, xarr, yarr)` para obtener un array que tome valores de `xarr` donde `cond` es `True` y de `yarr` donde es `False`. Verifica que el resultado es `[1.1, 2.2, 1.3, 1.4, 2.5]`.
- Crea una matriz aleatoria 4×4 con `rng.standard_normal((4, 4))` y usa `np.where` para reemplazar todos los valores positivos por 2 y los negativos por −2.

````python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])


resultado = np.where(cond, xarr, yarr)
print(resultado)

rng = np.random.default_rng(42)
matriz = rng.standard_normal((4, 4))
print("Matriz original:\n", matriz)

matriz_mod = np.where(matriz > 0, 2, -2)
print("Matriz modificada:\n", matriz_mod)

````

`np.where` también acepta escalares en lugar de arrays. Dado el array:

