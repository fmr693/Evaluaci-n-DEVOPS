# 10. Mostrar, gráficamente la relación entre el precio de venta y “GrLivArea”. ¿Cómo es? ¿Y la relación con “LotArea”?.


```python
import seaborn as sns
import matplotlib.pyplot as plt

variables = ['SalePrice', 'GrLivArea', 'LotArea']
datos = df[variables].dropna()

plt.figure(figsize=(8,6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=datos)
plt.title('Relación entre SalePrice y GrLivArea')
plt.xlabel('Superficie habitable (GrLivArea)')
plt.ylabel('Precio de venta (SalePrice)')
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(x='LotArea', y='SalePrice', data=datos)
plt.title('Relación entre SalePrice y LotArea')
plt.xlabel('Tamaño del lote (LotArea)')
plt.ylabel('Precio de venta (SalePrice)')
plt.show()
```
